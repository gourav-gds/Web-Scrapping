```python
def read_constituency_page(name, link):
    uClient = ureq(link)
    page_html = uClient.read()
    uClient.close()
    
    soup2 = BeautifulSoup(page_html, 'html.parser')

    #extracting all the tables
    results2 = soup2.find_all('table', attrs = {'class': 'wikitable'})
    
    #initialition required in case table we are looking for is not present
    party_votes = ''
    for table in results2:
        if table.find('a', attrs = {'title': '2015 Delhi Legislative Assembly election'}):
            party_votes = table          #if True, the following table gets assigned
    
    if len(party_votes) == 0:             #Deoli has no 2015 data, hence this check
        return name, np.nan, np.nan, np.nan
    trs = party_votes.find_all('tr', attrs = {'class': 'vcard'})
    
    
    #initializing the values; these variables will store the percentage votes
    others = 0
    AAP = np.nan
    BJP = np.nan         #BJP didnot contest from a certain place, it was returning error; hence everthing is intialized
    INC = np.nan
    
    #matching the party name
    for tr in trs:
        party = tr.find_all('td')[1].text
        if party[:-1] == 'BJP':
            BJP = tr.find_all('td')[4].text[:-1]       
        elif party[:-1] == 'AAP':
            AAP = tr.find_all('td')[4].text[:-1]
        elif party[:-1] == 'INC':
            INC = tr.find_all('td')[4].text[:-1]
        else:
            others += float(tr.find_all('td')[4].text)
    return name, AAP, BJP, INC, round(others, 2)
```

```python
records = []
for i in range(1,71):
    c = constituency.find_all('tr')[i]                         #checks each constituency 1 by 1
    name = c.find('a').text
    c.find('a')['href']                                        #link to the constituency page
    link = 'https://en.wikipedia.org' + c.find('a')['href']
    t = read_constituency_page(name, link)
    records.append((t))
```
